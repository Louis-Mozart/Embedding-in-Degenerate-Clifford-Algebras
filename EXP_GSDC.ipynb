## Results WN18-RR


```python
data_WN18 = pd.read_csv('GSDC_WN18RR.csv') # This file comes from the run_Decal_GSDC.py code with WN18RR as arguments
data_WN18.head()
```

```python
#top 5 best triples for WN18 on test data
time = 'Test' #available options ['Train','Test', 'Val']
p = show_performance(data_WN18, time, 'WN18RR')
latex = 'False'
k = 5
top5WN18 = p.topk_triples(latex,k)
top5WN18
```
